```python
trans_list = [('v1','t1'),('v2','t2'),('v3','t1')]
```

```python
trans_list
```

```python
d_trans = {}
for item in trans_list:
    print(f"key: {item[1]} and value is {item[0]}")
    if item[1] in d_trans:
        d_trans[item[1]].append(item[0])
    else:
        d_trans[item[1]] = [item[0]]
```
